Norimų stulpelių pervadinimas:

```python
zuvys_kartu.rename(
    columns={"Žuvų išteklių būklės indeksasi": "Žuvų būklės indeksas",
    "Biomasė (kg)i": "Biomasė(kg)",
    "Gausumas (vnt.)i": "Gausumas(vnt)",
    "Plėšrių žuvų gausumas (%)i": "Plėšrių žuvų gausumas(%)",
    "Plėšrių žuvų biomasė (%)i": "Plėšrių žuvų biomasė(%)"},
    inplace=True,)
```

Nereikalingo paskutinio stulpelio pašalinimas:

```python
zuvys_kartu.drop(columns=['Unnamed: 10'], inplace=True)
```

Indekso pakeitimas, kad elementų skaičiavimas prasidėtų nuo 1:

```python
zuvys_kartu['index'] = list(range(1, 63))
zuvys_kartu.set_index('index', inplace=True)
zuvys_kartu.head(10)
```

Kadangi nėra tuščių, nulinių ar vienodų reikšmių, keičiu "-" reikšmes į "0", kad galėčiau naudoti stulpelius skaičiavimams:

```python
zuvys_kartu = zuvys_kartu.replace(['-'], '0')
zuvys_kartu.head(10)
```

Patikrinu blogiausios būklės telkinio pagal būklės indekso reikšmę informaciją:

Patikrinu, kiek yra kokių būklės indekso reikšmių, rūšiuojant nuo didžiausio skaičiaus(1):

```python
zuvys_kartu['Žuvų būklės indeksas'].value_counts(sort=True)
```

Kadangi matau, kad net 9 telkiniai turi max vertinimą, tai .idxmax nepadės, nes rodys tik vieną(pirmą tinkamą) vandens telkinį.
Todėl ieškomi ir atvaizduojami visi variantai su max indeksu:

```python
zuvys_kartu.loc[zuvys_kartu['Žuvų būklės indeksas']==zuvys_kartu['Žuvų būklės indeksas'].max()].T
```

Taip pat patikrinama kiek telkinių ir kokioje savivaldybėje yra geriausios būklės(kai indeksas 1):

```python
tvarkom = zuvys_kartu.groupby(['Žuvų būklės indeksas'])['Savivaldybė'].value_counts()[1]
tvarkom
```

Grupuojant galima sužinoti kiek telkinių kokioje savivaldybėje buvo ištirta iš viso:

```python
vietove = zuvys_kartu.groupby('Savivaldybė')
vietove
```

```python
kokiu_daugiausia_vietu = vietove.count()
kokiu_daugiausia_vietu.head(24)
```

Kadangi norėsiu palyginti kokią dalį plėšri žuvis sudaro palyginus su visu sugautų žuvų kiekiu, 
tai plėšrių žuvų biomasę procentais reikia paversti į kilogramus:

```python
biomase_kg = 100
zuvys_kartu['Plėšrios(kg)'] = (zuvys_kartu['Biomasė(kg)'] * zuvys_kartu['Plėšrių žuvų biomasė(%)'] / biomase_kg).round(3)
zuvys_kartu.head(10)
```
